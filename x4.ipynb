```python
class x4():
    #initializes the connect 4 (x4) class
    columns = 7
    rows = 6
    
    def __init__(self):
        #creates blank spaces for i in columns, and j in rows aka creates a blank 7x6 board that will be filled later
        self.board = [[' ' for i in range(x4.columns)] for j in range(x4.rows)]
        
    def print_board(self):
        template = " "
        for i in self.board:
            template += '--' * 15 + '\n'
            #creates 15 double-dashes and then a line break used to seperate the rows of the board
            for j in i:
                template += f' | {j}'
            #creates a hidden {j} dictionary which populates between the '|' in the rows of the board
            template += ' | \n'
        #sets the right-most border of the board
        template += '--' * 15
        #sets the bottom of the board
        print(template)
        
    def drop(self, col, team:str):
        if col:
            if col.isdigit():
                col=int(col)
                #assigns drop value to the intended column
                if col <= 7 and col > 0:
                    col -=1
                    #helps allign the piece with the correct column because python indexes at '0' 
                    if self.board[0][col] == ' ':
                        #checks to make sure the intended column has an empty spot
                        for i in range(5,-1,-1):
                            #5 is the bottom row and 0 is the top row, we increment by -1 to ensure we fill the bottom of the board first
                            if self.board[i][col] == ' ':
                                self.board[i][col] = team
                                break
                    else: print('Column is full, lose a turn.') #if entry in a full column
                else: print('Invalid column, lose a turn.') #if entry not 1-to-7
            else: print('That is not a number, lose a turn.') #if entry not a number
        else: print ('Next time, type a number 1-to-7; lose a turn.')#if entry blank field
            
    def win(self, team:str):
    #checks the board for win conditions
        for i in range(x4.columns - 3):
            for j in range(x4.rows):
                if self.board[j][i] == team and self.board[j][i+1]== team and self.board[j][i+2] == team and self.board[j][i+3]== team:
                    return True #checks for four-in-a-row horizontally 
        
        for i in range(x4.columns):
            for j in range(x4.rows - 3):
                if self.board[j][i] == team and self.board[j+1][i]== team and self.board[j+2][i] == team and self.board[j+3][i]== team:
                    return True #checks for four-in-a-row vertically
                
        for i in range(x4.columns - 4, x4.columns):
            for j in range(x4.rows - 3):
                 if self.board[j][i] == team and self.board[j+1][i-1]== team and self.board[j+2][i-2] == team and self.board[j+3][i-3]== team:
                    return True #checks for four-in-a-row diagonally to the left
                
        for i in range(x4.columns - 3):
            for j in range(x4.rows - 3):
                 if self.board[j][i] == team and self.board[j+1][i+1]== team and self.board[j+2][i+2] == team and self.board[j+3][i+3]== team:
                    return True #checks for four-in-a-row diagonally to the right
                
    def tie(self):
        #checks to see if board is full a.k.a. a tie game
        for i in self.board:
            for j in i:
                if j == ' ':
                    return False
            return True
        
if __name__ == '__main__':
    board = x4()
    board.print_board()
    
    team1, team2 = "", ""
    while len(team1)!=1:
        team1 = input('Enter a letter as your token, player one: ')
    while len(team2) != 1 and team2 != team1:
        team2 = input('Enter a letter as your token, player two: ')
        if team2 == team1:
            print('Error, choose a different name.')
            
    while not board.win(team1) and not board.win(team2):
        col = input(f'Player one ({team1}) it\'s your turn, where do you want to drop your piece (1-7) ')
        board.drop(col, team1)
        board.print_board()
        
        if board.win(team1):
            print(f'You win, ({team1})!')
            quit()
            
        if board.tie():
            print('It\'\s a tie!')
            quit()
            
            
        col = input(f'Player two ({team2}) it\'s your turn, where do you want to drop your piece (1-7) ')
        board.drop(col, team2)
        board.print_board()
        
        if board.win(team2):
            print(f'You win, ({team2})!')
            quit()
            
        if board.tie():
            print('It\'\s a tie!')
            quit()
```
